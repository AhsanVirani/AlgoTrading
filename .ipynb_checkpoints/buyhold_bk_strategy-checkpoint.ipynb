```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
plt.style.use("seaborn")
```

```python
df = yf.download("TSLA", start="2017-01-01", end="2020-04-30")
```

```python
df['Close'].plot(figsize=(12,8), title='TSLA/USD', fontsize=10)
```

```python
# Log returns
df['returns'] = np.log(df['Close'].div(df['Close'].shift(1)))
```

```python
##########
# Strategy
##########
```

```python
# Drop all NA's
df.dropna(inplace=True)
```

```python
# Histogram
df['returns'].hist(bins=100, figsize=(12,8))
plt.title("TESLA/USD")
plt.show()
```

```python
# Cumulative returns
df['returns'].sum()
```

```python
# Investing $1 in Tesla at start till end date return $3.7
np.exp(df['returns'].sum())
```

```python
# Confirming 
df['Close'][-1] / df['Close'][0]
```

```python
df['returns'].cumsum().apply(np.exp)
```

```python
df['creturns'] = df['returns'].cumsum().apply(np.exp)
```

```python
# Plotting returns
df['returns'].plot(figsize=(12,8), title='TSLA/USD', fontsize=10)
```

```python
# plotting cum returns
df['creturns'].plot(figsize=(12,8), title='TSLA/USD', fontsize=10)
```
